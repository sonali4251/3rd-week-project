```python
# installation of Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from pandas import DataFrame
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import multilabel_confusion_matrix
from imblearn.over_sampling import SMOTE
from sklearn import preprocessing
import scipy.stats as stats
```

```python
#importing our data file "glass.csv" into a pandas dataframe
url = "https://raw.githubusercontent.com/dsrscientist/dataset3/main/glass.csv"   
names = ['RI','Na','Mg','Al','Si','K','Ca','Ba','Fe','Type']
glass_df = pd.read_csv(url, names=names)
```

# Since, there are no missing values in the dataset, there is no treatment needed.

In our case, the dependent variable (y) is "Type" which has 7 categories and the remaining variables are independent. We want to classify the glass type based on the independent features.

```python
#Separating the data into dependent and independent variables.
glass_X = glass_df.drop('Type', axis = 1)
glass_X.head()
```

```python
glass_Y = glass_df['Type']
glass_Y.head()
```

# Data Visualization

```python
#plotting a histogram of each independent feature for visualization
cols = ['RI', 'Na', 'Mg', 'Al', 'Si', 'K', 'Ca', 'Ba', 'Fe']
fig,ax = plt.subplots(3,3, figsize=(16, 12))
ax = ax.flatten()
i = 0
for col in cols:
    skew = glass_df[col].skew()
    sns.distplot(glass_df[col], ax = ax[i], fit= stats.norm, kde=False, label='Skew = %.3f' %(skew))
    ax[i].legend(loc='best')
    i += 1
plt.show()
```

Interpretation of Graphs : None of the features are normally distributed and some have outliers.Also, Outlier treatment may be done to check impact on classification

```python
#Scatter Matrix
pd.plotting.scatter_matrix(glass_df.iloc[:,:-1], c=glass_df.iloc[:,-1], figsize=(20, 20), marker='o')
plt.legend(glass_df["Type"].unique())
plt.show()
```

```python
# Checking for outliers
glass_df.iloc[:,:-1].boxplot(figsize=(12,6))
plt.show()
```

```python
#Outlier Treatment
numerical_features=['RI', 'Na', 'Mg', 'Al','Si','K','Ca','Ba','Fe']
for cols in numerical_features:
    Q1 = glass_X[cols].quantile(0.25)
    Q3 = glass_X[cols].quantile(0.75)
    IQR = Q3 - Q1  
    Lower_Whisker = Q1-1.5*IQR
    Upper_Whisker = Q3+1.5*IQR

    glass_X[cols] = (np.where(glass_X[cols] >= Upper_Whisker, Upper_Whisker, glass_X[cols]))
    glass_X[cols] = (np.where(glass_X[cols] <= Lower_Whisker, Lower_Whisker, glass_X[cols]))
```

```python
#Checking if the outliers are treated
glass_X.boxplot(figsize=(12,6))
plt.show()
```

```python
#Checking for correlation between independent variables
corr = glass_X.corr()
#Plot figsize
fig, ax = plt.subplots(figsize=(9, 7))
#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f")
#Apply xticks
plt.xticks(range(len(corr.columns)), corr.columns);
#Apply yticks
plt.yticks(range(len(corr.columns)), corr.columns)
#show plot
plt.show()
```

# Since there are 9 independent features out of which some are correlated and only 214 observations, we used Principal Component Analysis to prevent the data from overfitting and also reducing multicollinearity.

# Principal Component Analysis
Principal Component Analysis is a dimensionality reduction technique, connected with explaining the variance-covariance structure of a set of variables through a few linear combinations of these variables. There is as much information in the k Principal Components as there is in the original p variables and hence, the 'k' Principal Components can replace these original variables. Thus, the dimension of the data is reduced.

```python
# Reducing the dimensionality of our dataset using PCA
pca = PCA()
pca.fit(glass_X)
```

```python
pca.n_components_
```

```python
#To find the variance explained by different components
DataFrame(pca.explained_variance_.round(2), index=["P" + str(i) for i in range(1,10)], columns=["Explained Variance"]).T
```

```python
#Proportion of variance explained by individual components
DataFrame(pca.explained_variance_ratio_.round(2), index = ["P" + str(i) for i in range(1,10)], 
          columns=["Explained Variance Ratio"]).T
```

```python
# Since, first three components explain 90% of the total variation in the dataset, we choose them as our Principal Components.
pca = PCA(n_components = 3)
glass_X_pca = pca.fit_transform(glass_X)
```

```python
#Eigenvectors of the 3 Principal Components
pca.components_ 
```

```python
#Linear Combination
df = pd.DataFrame(pca.components_, columns=list(glass_X.columns))
df.head()
```

```python
#Explained Variance of 3 individual Principal Components
pca.explained_variance_ratio_ 
```

```python
print(sum(pca.explained_variance_ratio_))
```

# Total proportion of variance explained by the first 3 components is 90.459%

```python
#Splitting into train and test
X_train, X_test, y_train, y_test = train_test_split(glass_X_pca, glass_Y, test_size=1/5, random_state=5)
```

```python
#K Nearest Neighbours Classifier
test_scores = []
train_scores = []

# we will be testing 14 different values for k, starting with 1 and ending before 15
for i in range(1,15):

    knn = KNeighborsClassifier(i)
    knn.fit(X_train,y_train)
    
    train_scores.append(knn.score(X_train,y_train))
    test_scores.append(knn.score(X_test,y_test))
```

```python
max_train_score = max(train_scores)
train_scores_ind = [i for i, v in enumerate(train_scores) if v == max_train_score]
print('Max train score {} % and k = {}'.format(max_train_score*100,list(map(lambda x: x+1, train_scores_ind))))
```

```python
# This score comes from testing on the datapoints that were split in the beginning to be used for testing solely
max_test_score = max(test_scores)
test_scores_ind = [i for i, v in enumerate(test_scores) if v == max_test_score]
print('Max test score {} % and k = {}'.format(max_test_score*100,list(map(lambda x: x+1, test_scores_ind))))
```

```python
y_test.value_counts()
```

The accuracy of the KNN model comes out to be 74.418%.

```python
#Random Forest Classifier
clf=RandomForestClassifier(random_state = 45)
clf.fit(X_train,y_train)
y_preds=clf.predict(X_test)
mat=confusion_matrix(y_test,y_preds)
print(mat)
print(sns.heatmap(mat,annot=True,cmap='bwr',linewidths=.5))
acc=accuracy_score(y_test,y_preds)
print("Accuracy: ",acc*100,"%")
```

The accuracy of the Random Forest Classifier model comes out to be 65.1163%

# SMOTE
Oversampling Since, our model's dependent variable - Type of Glass, has an unequal distribution, our model might benefit from an oversampling technique called SMOTE. SMOTE will even out the difference by inserting synthetically produced rows of data which fall in between the already existing minority dependent variable.

```python
#As we can see, a couple of our types of glass have many more values than the rest. SMOTE will fill in the difference.
sm = SMOTE(random_state=42)
x_res, y_res = sm.fit_resample(X_norm, glass_Y)
y_res_df = pd.DataFrame(y_res)
ax = y_res_df.Type.value_counts().plot(kind="bar")
ax.set(xlabel="Type of Glass", ylabel="Count", title = "After SMOTE")
```

# CONCLUSION
Our aim was to classify the glass types correctly into 7 categories. There were 9 independent variables. After cleaning the data, that is, treating the outliers using IQR method, we visualized the data using histograms.
To reduce the dimensionality of the data, we used Principal Component Analysis which gave us 3 Principal Components that explained 90.459% of the total variation in the data.
To improve the fit of the model, we normalized the whole data and since, the distribution of glass types was highly imbalanced, we used an over-sampling technique- SMOTE. After this, we again used Principal Component Analaysis that reduced the dimensions to 5 components. These components explained 94.015% of total variation in the newly transformed dataset.
Now to classify the glass types, we used two Machine learning algorithms - KNN and Random Forest that classified the glass types with an accuracy of 74.4180% and 65.1163% before normalising and gave an accuracy of 80.04% and 91.304% after normalising respectively.
So, we conclude that after normalising and using SMOTE technique, Random forest proves to be a better fit for classifying the data. Thus, PCA reduces the dimensions of the data and also improves the algorithm's performance.
